# Cleaning a Dataset....

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.set_index("Passenger",inplace=True)
data
```

```python
newdata=data.fillna(0)
newdata
```

```python
newdata.isnull().sum()
```

```python
newdata=data.fillna({'Age':'Nothing','Cabin':0})
newdata
```

```python
newdata=data.fillna(method='ffill')
newdata
```

```python
newdata=data.interpolate()
newdata
```

```python
newdata_1=newdata.drop(['Cabin','Parch','Ticket'], axis=1)
newdata_1
```

```python
newdata_1.isnull().sum()
```

```python
newdata_2=newdata_1.fillna({'Embarked':'S'})
newdata_2
```

```python
newdata_2.isnull().sum()
```

```python
Q1=newdata_2.Fare.quantile(0.25)
Q3=newdata_2.Fare.quantile(0.75)
Q1,Q3
```

```python
IQR=(Q3-Q1)/2
IQR
```

```python
LL=Q1-1.5*IQR
UL=Q3-1.5*IQR
LL,UL
```

```python
newdata_2[(newdata_2.Fare<LL)|(newdata_2.Fare>UL)]
```
